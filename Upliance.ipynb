```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
user_details = pd.read_csv('UserDetails.csv')
cooking_sessions = pd.read_csv('CookingSessions.csv')
order_details = pd.read_csv('OrderDetails.csv')
user_details

```

```python
user_details['Registration Date'] = pd.to_datetime(user_details['Registration Date'])
cooking_sessions['Session Start'] = pd.to_datetime(cooking_sessions['Session Start'])
cooking_sessions['Session End'] = pd.to_datetime(cooking_sessions['Session End'])
order_details['Order Date'] = pd.to_datetime(order_details['Order Date'])


```

```python
order_details['Rating']=order_details['Rating'].fillna(0.0)
order_details
```

```python
merged_data = order_details.merge(cooking_sessions, on=['Session ID', 'User ID', 'Dish Name', 'Meal Type'], how='left')
merged_data = merged_data.merge(user_details, on='User ID', how='left')

mergeall3data=merged_data.merge(order_details,on='User ID',how='left')
merged_data





```

```python
mergeall3data
```

```python
# relationship between cooking sessions and user orders
# Calculate the avg duration of cooking sessions for each dish
avg_duration_per_dish = merged_data.groupby('Dish Name')['Duration (mins)'].mean().reset_index()
avg_duration_per_dish
```

```python
user_analysis = merged_data.groupby('User ID').agg({'Duration (mins)': 'sum','Total Orders': 'first'}).reset_index()
user_analysis
```

```python
# Identify popular dishes
popular_dishes = merged_data['Dish Name'].value_counts().reset_index()
popular_dishes.columns = ['Dish Name', 'Order Count']
popular_dishes.columns
```

```python
#demographic factors that influence user behavior
#avg number of orders per user by age group
user_details['Age Group'] = pd.cut(user_details['Age'], bins=[20, 30, 40, 50], labels=['20-30', '30-40', '40-50'])
avg_orders_per_age_group = user_details.groupby('Age Group')['Total Orders'].mean().reset_index()
avg_orders_per_age_group
```

```python
# Create visualizations to showcase key insights
plt.figure(figsize=(12, 6))
sns.barplot(x='Dish Name', y='Order Count', data=popular_dishes)
plt.title('Popular Dishes')
plt.xticks(rotation=45)
plt.show()
plt.figure(figsize=(12, 6))
sns.barplot(x='Dish Name', y='Duration (mins)', data=avg_duration_per_dish)
plt.title('Average Duration of Cooking Sessions per Dish')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='Age Group', y='Total Orders', data=avg_orders_per_age_group)
plt.title('Average Number of Orders per User by Age Group')
plt.show()





```

```python

user_details = pd.read_csv('UserDetails.csv')
user_details['Registration Date'] = pd.to_datetime(user_details['Registration Date'])
#avg number of orders per user by age group
user_details['Age Group'] = pd.cut(user_details['Age'], bins=[20, 30, 40, 50], labels=['20-30', '30-40', '40-50'])
avg_orders_per_age_group = user_details.groupby('Age Group')['Total Orders'].mean().reset_index()
#pie chart for average number of orders per user by age group
plt.figure(figsize=(8, 8))
plt.pie(avg_orders_per_age_group['Total Orders'], labels=avg_orders_per_age_group['Age Group'], autopct='%1.1f%%', startangle=140)
plt.title('Average Number of Orders per User by Age Group')
plt.show()
```

```python
order_details = pd.read_csv('OrderDetails.csv')
# Convert date columns to datetime
order_details['Order Date'] = pd.to_datetime(order_details['Order Date'])

plt.figure(figsize=(14, 8))
sns.histplot(data=order_details, x='Order Date', hue='Meal Type', multiple='dodge', bins=30, kde=False)
plt.title('Histogram of Meal Types on All Specific Dates')
plt.xlabel('Order Date')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
merged_data = pd.merge(cooking_sessions, user_details, on='User ID')
# Group by 'User ID' and calculate the total duration of cooking sessions and total orders
user_analysis = merged_data.groupby('User ID').agg({'Duration (mins)': 'sum','Total Orders': 'first'}).reset_index()
plt.figure(figsize=(12, 6))
colors = sns.color_palette('husl', len(user_analysis))
sns.barplot(x='User ID', y='Total Orders', data=user_analysis, palette=colors)
plt.title('Total Orders by User')
plt.xlabel('User ID')
plt.ylabel('Total Orders')
plt.show()
```
